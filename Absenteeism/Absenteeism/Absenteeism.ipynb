```python
import pandas as pd
```

```python
data = pd.read_csv("Absenteeism_data.csv")
```

```python
data.head()
```

```python
# Copying data set
df = data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
# remove irrelevant data
df = df.drop(['ID'], axis = 1)
```

```python
sorted(df['Reason for Absence'].unique())
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
```

```python
# drop reason 0, which is an unknown reason
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
reason_columns
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
# group the reasons for absence
reason_type1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type4 = reason_columns.loc[:, 22:28].max(axis = 1)
```

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis = 1)
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
# reorder columns
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
                          'Daily Work Load Average', 'Body Mass Index', 'Education',
                          'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

```python
# make a checkpoint
df_2 = df.copy()
```

```python
df_2['Date'] = pd.to_datetime(df_2['Date'], format = '%d/%m/%Y')
```

```python
df_2['Date']
```

```python
type(df_2['Date'][0])
```

```python
df_2.shape
```

```python
# extract the month value

list_months = []
for i in range(700):
    list_months.append(df_2['Date'][i].month)
```

```python
len(list_months)
```

```python
df_2['Month Value'] = list_months
```

```python
df_2.head()
```

```python
# extract day of the week

def date_to_weekday(date_value):
    return date_value.weekday()

df_2['Day of the Week'] = df_2['Date'].apply(date_to_weekday)
```

```python
df_2.head()
```

```python
# drop Date column
df_2 = df_2.drop(columns = ['Date'])
```

```python
df_2.columns.values
```

```python
column_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                'Month Value','Day of the Week', 'Transportation Expense', 
                'Distance to Work', 'Age', 'Daily Work Load Average', 
                'Body Mass Index', 'Education','Children', 'Pets', 
                'Absenteeism Time in Hours']

df_2 = df_2[column_names]
```
